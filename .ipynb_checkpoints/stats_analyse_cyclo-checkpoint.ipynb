## 2/ Préparation des données

### 2.1/ Relation entre le nombre d'accident et l'atmosphère
H0 : Il n'y a pas de lien entre l'atmosphère et le nombre d'accident.
H1 : La qualité de la météo va influencer les accidents.

```python
dataset1 = acc_velos[ ["atm", "nb_acc"] ]
dataset1 = dataset1.groupby(by="atm", as_index=False)["nb_acc"].sum()
dataset1
```

```python
seaborn.barplot(x=dataset1["atm"], y=dataset1["nb_acc"])
plt.title("Nombre d'accident en fonction de la qualité de la météo")
plt.show()
```

### 2.2/ Relation entre le type d'intersection et le nombre d'accident
H0 : Il n'y a pas de lien entre le type d'intersection et le nombre d'accident.
H1 : Le type d'intersection va influencer les accidents.

```python
dataset2 = acc_velos[ ["int", "nb_acc"] ]
dataset2 = dataset2.groupby(by="int", as_index=False)["nb_acc"].sum()
dataset2.head()
```

```python
seaborn.barplot(x=dataset2["int"], y=dataset2["nb_acc"])
plt.title("Nombre d'accident en fonction du type d'intersection")
plt.show()
```

Il semblerait que ni l'atmosphère ni le type d'intersection influence visuellement les accidents de vélos. On va donc charger la table des lieux, pour savoir si d'autres caractéristiques du lieu peuvent influencer le nombre d'accident. Actuellement, on apprend que les accidents ont lieu par temps chaud, en dehors des intersections.

## 3/ Relation entre le nombre d'accident et les caractéristiques des lieux

```python
## Charger la table des lieux
lieux = pandas.read_csv(".\\DONNEES\\lieux-2021.csv", encoding="utf-8", sep=";", decimal=".")
lieux.head()
```

```python
lieux.dtypes
```

```python
dataset2 = pandas.merge(left=acc_velos, right=lieux, on="Num_Acc", how="left")
dataset2.head()
```

```python
dataset2.dtypes
```

### 3.1/ Relation entre le nombre d'accident et le régime de circulation

```python
dataset3 = dataset2[ ["circ", "nb_acc"] ]
dataset3 = dataset3.groupby(by="circ", as_index=False)["nb_acc"].sum()
dataset3.head()
```

```python
seaborn.barplot(x=dataset3["circ"], y=dataset3["nb_acc"])
plt.title("Nombre d'accidents en fonction du type de circulation")
plt.show()
```

Quand on se réfère à la documentation, on voit que le 2 correspond à une circulation bidirectionnelle. Donc, on a plus d'accident dans ce genre de situation. On va voir, si, effectivement, ça a du sens pour un aménagement.

```python
## Le test de fischer s'effectue sur des tableaux de contingence.
## On crée un tableau de contingence :
res3 = scipy.stats.contingency.crosstab(dataset3)
res3
```

```python
## Puis on réalise le test exact de Fischer

scipy.stats.fisher_exact(res3, alternative='two-sided')
```
