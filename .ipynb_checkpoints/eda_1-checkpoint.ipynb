### Traitement du temps et de l'heure
On voit qu'on a un champ pour l'heure, et un 3 champs qui peuvent former une date.
Python possède une [classe date et datetime](https://docs.python.org/fr/3/library/datetime.html). 

**Intérêts de cette classe** :
- On peut faire des calculs de distance de temps (delta de temps).
- Regroupement par période (facilite le traitement du temps).

#### Créer un objet datetime
[convertir le champ hrmn en type time python](https://www.digitalocean.com/community/tutorials/python-string-to-datetime-strptime)
[Utiliser la documentation pour faire le traitement](https://docs.python.org/fr/3/library/datetime.html#strftime-and-strptime-behavior)

```python
## Mettre à jour le jeu de données en sélectionnant uniquement le département du Val-de-Marne :
acc = acc[ acc["dep"] == "94" ]
acc.head()
acc.shape ## (2493, 15)
```

```python
## Créer une colonne de datetime :
acc["datetime"] = ""
acc.shape ## (2493, 16)
```

```python
## Recréer les index :
acc.index = list( range(0, acc.shape[0]) )
acc.head()
```

Lors de la création du champ datetime dans la table, la boucle retournait une erreur au bout d'un moment à cause des index. La méthode .iloc() considérait qu'on essayait d'agrandir le df. Tous les index sont passés sauf le 72e. Du coup, il semblait que le soucis venait de la compréhension du 73e index pour la fonction. On a donc transformé tous les index, allant de 0 à la taille du jeu de données. On a relancé la boucle pour créer un champ datetime, et ça marche !

```python
## pour chaque enregistrement, associer sa datetime à partir des champs existants
## On créera une date en français dans la mesure du possible :

for i, r in acc.iterrows() :
    ## séparer la date et l'heure :
    h = r["hrmn"].split(":")[0] ## récupérer l'heure
    mn = r["hrmn"].split(":")[1] ## récupérer les minutes
    
    dtm = f"{r['jour']}/{r['mois']}/{r['an']} {h}:{mn}" ## puis assembler pour créer un objet datetime
    res = datetime.datetime.strptime(dtm, "%d/%m/%Y %H:%M") ## créer un objet datetime
    
    ## ajouter le résultat au df à la ligne correspondante :
    acc.iloc[i, 15] = res
    
acc.head()
```

```python
acc.tail()
```

On a désormais un champ avec le datetime et donc on peut y exploiter toutes les méthodes de datetime, notamment le delta de temps. #### On peut créer une colonne de geometry pour geopandas => utiliser shapely

Les accidents sont localisés en WGS84, il faudra donc :
1. Créer un point depuis geopandas,
2. Créer un nouveau champ de géométrie, qui va contenir le point reprojetée en 3949,
3. Puis reprojeter le point,
4. Supprimer le premier champ de géométrie crée, et utiliser le nouveau comme référence,
5. Renommer les champs pour qu'ils soient compréhensibles,
6. Extraire les coordonnées en 3949 Lambert-93

```python
acc.dtypes
```

```python
# Utilisation de la méthode "apply" pour appliquer la méthode "astype" à chaque élément de la colonne
acc['long'] = acc['long'].apply(lambda x: float(x))
acc['lat'] = acc['lat'].apply(lambda x: float(x))

## En fait, les opérateurs de comparaisons ne fonctionnent pas très bien avec pandas.
## En utilisant == pour tester le type, on a une ValueError.
## Pour palier à ce problème, on utilisera la méthode apply() pour convertir.
## Une telle conversion suppose qu'on connaît bien le type.

acc = geopandas.GeoDataFrame(acc, geometry=geopandas.points_from_xy(acc.long, acc.lat))
acc.head()

## A la naissance les POINT geopandas, sans précision, n'ont pas CRS.
## Il faut donc en attribuer un :
acc["geometry"] = acc["geometry"].set_crs(epsg=4326)

## Ensuite, il faut créer un nouveau champ qui va contenir la géométrie reprojetée en 3949 :
acc["geomLam93"] = acc["geometry"].to_crs(epsg=3949)
```

```python
## supprimer la géométrie WGS84 :
acc.drop(["geometry"], axis=1, inplace=True)
```

```python
acc.head()
```

```python
acc.rename(columns={"geomLam93" : "geometry"}, inplace=True)
acc.head()
```

```python
## La représentation consiste en un croisement entre deux variables ou plus
## On pourrait commencer par afficher le nombre d'accident pour chaque communes de GPSEA

insee_gpsea = ["94002", "94028", "94011", "94044", 
               "94019", "94055", "94053", "94071", 
               "94004", "94075", "94048", "94047",
               "94056", "94070", "94060", "94059"]
acc_gpsea = acc[ acc["com"].isin(insee_gpsea) ]

plt.figure(figsize=(16, 6), dpi=100)
seaborn.countplot(data=acc_gpsea,
                 x="com")
plt.show()
```

Mais qu'en est-il des variations mensuelles ?

```python
## On va essayer de regarder les variations mensuelles :
acc["datetime"][0].month ## récupère le mois

## On va pouvoir représenter les données mensuellement par un lineplot :
insee_gpsea = ["94002", "94028", "94011", "94044", 
               "94019", "94055", "94053", "94071", 
               "94004", "94075", "94048", "94047",
               "94056", "94070", "94060", "94059"]
acc_gpsea = acc[ acc["com"].isin(insee_gpsea) ]

## Tant qu'on ré-index pas notre tableau, il y aura des problèmes de lecture des indice pour le iloc :
acc_gpsea.index = list(range(0, acc_gpsea.shape[0]))
acc_gpsea["nb_acc"] = 1
acc_gpsea["mois"] = ""

for i, r in acc_gpsea.iterrows() :
    acc_gpsea.iloc[i, 13] = acc_gpsea.iloc[i, 9].month
```

```python
 ## Puis on peut faire un group by et sommer :
temp = acc_gpsea[ ["com", "nb_acc", "mois"] ]
temp = temp.groupby(by=["com", "mois"])["nb_acc"].sum().reset_index()

temp.head() ## là c'est une série.
```

```python
## Ensuite on peut afficher le lineplot :
plt.figure(figsize=(16, 6), dpi=100)

acc_trends = seaborn.lineplot(data=temp, x="mois", y="nb_acc", hue="com")
## Replacer la légende :
seaborn.move_legend(obj=acc_trends, 
                    loc="upper left",
                    ncols=2,
                    title="Communes")

## source pour l'association des codes INSEE avec les codes insee : https://fr.wikipedia.org/wiki/%C3%89tablissement_public_territorial_Grand_Paris_Sud_Est_Avenir
plt.show()
```

Pour nettoyer cette table, il suffit de récupérer les identifiants des accidents du 94 issu de la table caracteristiques.

```python
lieux = lieux[ lieux["Num_Acc"].isin(acc["Num_Acc"]) ]
lieux.shape ## (2493, 11)
```

```python
## Pour chaque commune montrer dans quelle situation de surface de sol les accidents sont le plus représentés :
acc_lieux = acc.set_index("Num_Acc").join(lieux.set_index("Num_Acc"), how="left")
acc_lieux.head()
```

```python
acc_lieux.isna().sum()
```

```python
acc_gpsea = acc_lieux[ acc_lieux["com"].isin(insee_gpsea)]

plt.figure(figsize=(16, 6), dpi=100)

seaborn.countplot(data=acc_gpsea,
                  x="com", hue="surf")
plt.show()
```

On voit que :
- Principalement, les accidents ont lieux sur surface sèche.
- Ensuite vient les surfaces mouillées.

Bien qu'on n'établissent pas là de corrélation ou d'absence de corrélation, le résultat est intéressant. Une mauvaise qualité de la route ne varie pas la même manière que les accidents (majoritairement des collisions sur les côtés et à l'arrière). Cette observation est réalisée sur l'ensemble de l'année !
